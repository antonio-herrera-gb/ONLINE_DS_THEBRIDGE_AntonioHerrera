### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
top_cities = df.nlargest(5, "population_total")

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Grfico de barras
axs[0, 0].bar(top_cities["city"], top_cities["population_total"] / 1000, color=["blue", "gray", "olive", "brown", "maroon"], alpha=0.7)
for i, val in enumerate(top_cities["population_total"] / 1000):
    axs[0, 0].text(i, val + 100, f"{val:.3f}", ha='center', fontsize=8)
axs[0, 0].set_title("Ciudades ms pobladas de California", fontsize=12)
axs[0, 0].set_ylabel("Habitantes (miles)")

# Histograma del area
axs[0, 1].hist(df["area_land_sq_mi"], bins=50, alpha=0.8, color="steelblue", label="Hist Area")
axs[0, 1].set_title("Superficie (km)", fontsize=12)
axs[0, 1].set_ylabel("Nmero de ciudades")
axs[0, 1].set_xlabel("Superficie ($km^2$)")
axs[0, 1].legend()

# Scatter de poblacin vs rea
axs[1, 0].scatter(df["area_land_sq_mi"], df["population_total"] / 1000, alpha=0.6, color="green")
axs[1, 0].set_title("Poblacin por kilmetro cuadrado de agua", fontsize=12)
axs[1, 0].set_xlabel("rea de agua en la ciudad ($km^2$)")
axs[1, 0].set_ylabel("Poblacin (miles habitantes)")
axs[1, 0].annotate("Outliers", xy=(455, 3900), xytext=(40, 3500),
                   arrowprops=dict(facecolor='black', arrowstyle="->"), fontsize=10)

# Boxplot
axs[1, 1].boxplot(df["latd"]*100, vert=False, patch_artist=True, boxprops=dict(facecolor="red"))
axs[1, 1].set_title("Diagrama de caja de la Altitud de las ciudades de California", fontsize=12)
axs[1, 1].set_xlabel("Altitud (m)")
```
